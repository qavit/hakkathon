```python
# Load 四縣腔詞條詞目文字.ods
import pandas as pd

df = pd.read_excel('/content/四縣腔詞條詞目文字.ods', engine='odf')
df
```

```python
# prompt: In the column '例句' of df，the example sentences are structured in the following pattern: '例：<hak_sectence_1> (<zh_sectence_1>)'. If an entry has multipule sentences, the pattern repeats. Please parse and save example sentences as `example_sentences.json`. The keys are `dict_id` ('序號' of df), `hak_ex`, `zh_ex`. Count the number of sentence pairs. Use tqdm in for loop.

from tqdm import tqdm
import json

example_sentences = []
sentence_pair_count = 0

for index, row in tqdm(df.iterrows(), total=df.shape[0]):
  dict_id = row['序號']
  if isinstance(row['例句'], str):
    examples = row['例句'].split('例：')
    for example in examples:
      if example:
        parts = example.split('(')
        if len(parts) == 2:
          hak_ex = parts[0].strip()
          zh_ex = parts[1].strip()[:-1]  # Remove the closing parenthesis
          example_sentences.append({
              'dict_id': dict_id,
              'hak_ex': hak_ex,
              'zh_ex': zh_ex
          })
          sentence_pair_count += 1

with open('example_sentences.json', 'w', encoding='utf-8') as f:
  json.dump(example_sentences, f, ensure_ascii=False, indent=2)

print(f"Total sentence pairs: {sentence_pair_count}")
```

```python
# prompt: calculate how many entries of example_sentences are problemtatic. Problem 1: does not contain '。' or '？' or '！' . Problem 2: contains full-width parentheses '（' or '）' and save them in a JSON file. Store the non-problematic entries in `filtered_sentences`. Print the number of both problematic and filtered sentences

problematic_sentences = []
filtered_sentences = []

for sentence in example_sentences:
  hak_ex = sentence['hak_ex']
  if '。' not in hak_ex and '？' not in hak_ex and '！' not in hak_ex:
    problematic_sentences.append(sentence)
  elif '（' in hak_ex or '）' in hak_ex:
    problematic_sentences.append(sentence)
  else:
    filtered_sentences.append(sentence)


with open('problematic_sentences.json', 'w', encoding='utf-8') as f:
  json.dump(problematic_sentences, f, ensure_ascii=False, indent=2)

print(f"Number of problematic sentences: {len(problematic_sentences)}")
print(f"Number of filtered sentences: {len(filtered_sentences)}")
```

```python
import re

def process_sentences(input_file, output_file):
    # Load the JSON data
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    processed_entries = []

    # Process each entry in the JSON
    for entry in tqdm(data, desc="Processing entries"):
        dict_id = entry["dict_id"]
        hak_ex = entry["hak_ex"]
        zh_ex = entry["zh_ex"]

        # Split hak_ex by '、' and check for parentheses
        hak_phrases = re.split(r'、', hak_ex)
        zh_default = zh_ex if '、' not in zh_ex else zh_ex.split('、')[-1]

        # Iterate through hak_phrases, keeping only items with matching parentheses or the last item
        for i, hak_phrase in enumerate(hak_phrases):
            match = re.search(r'（(.*?)）', hak_phrase)

            # Include only phrases with parentheses or the last phrase without parentheses
            if match or i == len(hak_phrases) - 1:
                hak_phrase_cleaned = re.sub(r'（.*?）', '', hak_phrase)
                zh_translation = match.group(1) if match else zh_default
                processed_entries.append({
                    "dict_id": dict_id,
                    "hak_ex": hak_phrase_cleaned,
                    "zh_ex": zh_translation
                })

    # Save the processed entries to the output file
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(processed_entries, f, ensure_ascii=False, indent=2)
```
